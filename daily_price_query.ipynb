```python
stock_str = "('" + "','".join(stocks)+ "')"
```

```python
query = f"""
    SELECT 
         *
    FROM
        crsp_a_stock.dsfhdr AS c
    WHERE
        c.htsymbol IN {stock_str}
"""

header = conn.raw_sql(query)

# query of the header file, we need to select the list we want from here. 
```

```python
cusips = header['hcusip']
```

```python
cusips_str = "('" + "','".join(cusips)+ "')"
```

```python
query = f"""
    SELECT 
        c.cusip,
        c.date,
        c.bid,
        c.ask,
        c.vol
    FROM
        crsp_a_stock.dsf AS c
    WHERE
        c.date > '2023-12-01' AND
        c.cusip IN {cusips_str}
"""

conn.raw_sql(query)
```
